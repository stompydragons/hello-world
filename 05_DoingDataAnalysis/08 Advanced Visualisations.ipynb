```python
#imports
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline

processed_data_path = os.path.join(os.getcwd(), 'data', 'processed')
write_train_path = os.path.join(processed_data_path, 'train.csv')
write_test_path = os.path.join(processed_data_path, 'test.csv')

train_df = pd.read_csv(write_train_path)
test_df = pd.read_csv(write_test_path)
df = pd.concat([train_df, test_df], axis=0)
```

# Advanced Visualizations

Unlike pandas high level plotting, in matplotlib, you need to create all of the components on your own. But the results are nicer!

```python
plt.hist(df.Age)
```

To hide all the matplotlib information added above, you can use the show() method.

```python
plt.hist(df.Age, bins=20, color='c')
plt.show()
```

```python
# add labels, use edgecolor to outline the bars, the default for this was changed in 2.0
plt.hist(df.Age, bins=20, color='c', edgecolor='black')
plt.title('Histogram: Age')
plt.xlabel('Bins')
plt.ylabel('Counts')
plt.show()
```

```python
# another way to create the same histogram, useful if you want to add subplots
# extract figure and axis from pyplot
f, ax = plt.subplots(figsize=(15,6))
ax.hist(df.Age, bins=20, color='c', edgecolor='black')
ax.set_title('Histogram: Age')
ax.set_xlabel('Bins')
ax.set_ylabel('Counts')
plt.show()
```

```python
# here we can add some subplots i.e. multiple plots in a single visualization
# this will create subplots that are 1 row, 2 columns
f, (ax1, ax2) = plt.subplots(1,2, figsize=(14,3))

# then separate parts can be created
ax1.hist(df.Fare, bins=20, color='c', edgecolor='black')
ax1.set_title('Histogram: Fare')
ax1.set_xlabel('Bins')
ax1.set_ylabel('Counts')

# and another subplot using ax2
ax2.hist(df.Age, bins=20, color='tomato', edgecolor='black')
ax2.set_title('Histogram: Age')
ax2.set_xlabel('Bins')
ax2.set_ylabel('Counts')

plt.show()
```

```python
# five subplots, mwah ha ha!
f, ax_arr = plt.subplots(3,2, figsize=(14,9))

# ax_arr is a two dimensional array, axes must be referred to with indexes
# Plot 1
ax_arr[0,0].hist(df.Fare, bins=20, color='c', edgecolor='black')
ax_arr[0,0].set_title('Histogram: Fare')
ax_arr[0,0].set_xlabel('Bins')
ax_arr[0,0].set_ylabel('Counts')

# Plot 2
ax_arr[0,1].hist(df.Age, bins=20, color='c', edgecolor='black')
ax_arr[0,1].set_title('Histogram: Age')
ax_arr[0,1].set_xlabel('Bins')
ax_arr[0,1].set_ylabel('Counts')

# Plot 3
ax_arr[1,0].boxplot(df.Fare.values)
ax_arr[1,0].set_title('Boxplot: Fare')
ax_arr[1,0].set_xlabel('Fare')
ax_arr[1,0].set_ylabel('Fare')

# Plot 4
ax_arr[1,1].boxplot(df.Age.values)
ax_arr[1,1].set_title('Boxplot: Age')
ax_arr[1,1].set_xlabel('Age')
ax_arr[1,1].set_ylabel('Age')

# Plot 5
ax_arr[2,0].scatter(df.Age, df.Fare, color='c', alpha=0.15)
ax_arr[2,0].set_title('Scatter plot: Age vs Fare')
ax_arr[2,0].set_xlabel('Age')
ax_arr[2,0].set_ylabel('Fare')

# to turn off an axis (subplot) you don't want to see
ax_arr[2,1].axis('off')

# to get rid of overlap issues, use tight_layout to add some padding
plt.tight_layout()

plt.show()
```

Lots of nice ideas here to get started with pyplot.
